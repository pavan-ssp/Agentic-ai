CBOW : [ Continous Bag of Words]
-----

Predict the target word from context words

Uses multiple context words to predict one target word

Out of vocabulary issue we will face it will generate output for vocabulary words only for example the words which are not in vocab we pass that will not process
and give output as outofvocabulary

example :
----------

"The cat sat on the mat"

CBOW Example:

Context words: ["The", "sat"]
Target word: "cat"



 Skip-gram :
 ----------

 Predict context words from a target word

 Uses one target word to predict multiple context words

 Out of vocabulary issue we will face it will generate output for vocabulary words only for example the words which are not in vocab we pass that will not process
 and give output as outofvocabulary



example :
----------

"The cat sat on the mat"



 Skip-gram Example:
 -------------------

Target word: "cat"
Context words: ["The", "sat"]

```python
# Import necessary libraries
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK tokenizer data (if not already downloaded)
nltk.download('punkt')
nltk.download('punkt_tab') # Download the punkt_tab resource

# Sample text corpus
text = "The quick brown fox jumps over the lazy dog. The dog barked at the fox."

# Tokenize sentence into words
tokenized_text = [word_tokenize(text.lower())]  # Convert text to lowercase and tokenize

# Train CBOW model (sg=0 for CBOW)
cbow_model = Word2Vec(sentences=tokenized_text, vector_size=50, window=2, min_count=1, sg=0)

# Train Skip-gram model (sg=1 for Skip-gram)
skipgram_model = Word2Vec(sentences=tokenized_text, vector_size=50, window=2, min_count=1, sg=1)

# Check vector representation of 'fox' in CBOW model
print("Vector representation of 'fox' in CBOW model:\n", cbow_model.wv['fox'])

# Check vector representation of 'fox' in Skip-gram model
print("\nVector representation of 'fox' in Skip-gram model:\n", skipgram_model.wv['fox'])

# Find similar words to 'fox' in CBOW model
print("\nWords similar to 'fox' in CBOW model:\n", cbow_model.wv.most_similar('fox'))

# Find similar words to 'fox' in Skip-gram model
print("\nWords similar to 'fox' in Skip-gram model:\n", skipgram_model.wv.most_similar('fox'))
```
